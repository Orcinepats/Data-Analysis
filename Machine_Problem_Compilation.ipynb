```python
def exponential(n, x):

    sum = 1.0
    for i in range(n, 0, -1):
        sum = 1 + x * sum / i
    print ("e^x =", sum)

# Input section
print("=============================================================")
print("===================== Machine Problem 1 =====================")
print("\nCalculate a series of inputs using Taylor Series")
n = int(input("\nInput value (n): "))
x = int(input("Input Value (x): "))

exponential(n, x)
```
